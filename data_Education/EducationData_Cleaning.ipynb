#Read in full census education data and clean up

```python
xls_file_full = pd.ExcelFile('nomis_2017_11_29_112403.xlsx.xls')
xls_file_full
df_full = xls_file_full.parse()
df_full_edit = df_full.drop(df_full.index[[0, 1, 2, 3, 4, 5, 6, 7]])
df_full_edit_2 = df_full_edit[:-2]
df_full_edit_2.columns._data[0] = 'Local Authority'
df_full_edit_2.columns._data[1] = 'All Categories'
df_full_edit_2.columns._data[2] = 'No Qualifications'
df_full_edit_2.columns._data[3] = 'Level 1 Qualifications'
df_full_edit_2.columns._data[4] = 'Level 2 Qualifications'
df_full_edit_2.columns._data[5] = 'Apprenticeship'
df_full_edit_2.columns._data[6] = 'Level 3 Qualifications'
df_full_edit_2.columns._data[7] = 'Level 4 Qualifications and above'
df_full_edit_2.columns._data[8] = 'Other Qualifications'
df_full_edit_2 = df_full_edit_2[['Local Authority', 'No Qualifications', 'Level 1 Qualifications', 'Level 2 Qualifications', 'Apprenticeship', 'Level 3 Qualifications', 'Level 4 Qualifications and above', 'Other Qualifications', 'All Categories']]
#df_full_edit_2['Local Authority'] = df_full_edit_2.Column1.str.replace('uacounty09:' , '')
#df_full_edit_2['Local Authority'].str.replace('uacounty09:','')
df_full_edit_2['Local Authority'].replace(regex=True,inplace=True,to_replace=r'uacounty09:',value=r'')
df_full_edit_2['Local Authority'].replace(regex=True,inplace=True,to_replace=r'ualad09:',value=r'')
df_full_edit_2['Local Authority'].replace(regex=True,inplace=True,to_replace=r'ccg2011:',value=r'')

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('EducationData.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df_full_edit_2.to_excel(writer, sheet_name='Sheet1')

#df_full_edit_2.to_csv("Education_Census_Data.csv")
df_full_edit_2
```

#Read in vets csv file and clean up
